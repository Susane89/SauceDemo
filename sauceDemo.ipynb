```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

def test_successful_purchase():
    # Инициализация драйвера Edge
    driver = webdriver.Edge()  # Используем браузер Edge
    
    try:
        # 1. Открыть сайт
        driver.get("https://www.saucedemo.com")
        
        # 2. Авторизация
        username_field = driver.find_element(By.ID, "user-name")
        password_field = driver.find_element(By.ID, "password")
        login_button = driver.find_element(By.ID, "login-button")
        
        username_field.send_keys("standard_user")
        password_field.send_keys("secret_sauce")
        login_button.click()
        time.sleep(2)  # Задержка после нажатия кнопки
        
        # Проверка успешного входа (переход на страницу товаров)
        WebDriverWait(driver, 10).until(
            EC.url_to_be("https://www.saucedemo.com/inventory.html")
        )
        
        # 3. Добавление товара в корзину
        add_to_cart_button = driver.find_element(By.ID, "add-to-cart-sauce-labs-backpack")
        add_to_cart_button.click()
        time.sleep(2)  # Задержка после нажатия кнопки
        
        # 4. Переход в корзину
        cart_icon = driver.find_element(By.CLASS_NAME, "shopping_cart_link")
        cart_icon.click()
        time.sleep(2)  # Задержка после нажатия кнопки
        
        # Проверка перехода на страницу корзины
        WebDriverWait(driver, 10).until(
            EC.url_to_be("https://www.saucedemo.com/cart.html")
        )
        
        # 5. Проверка наличия товара в корзине
        item_in_cart = driver.find_element(By.CLASS_NAME, "inventory_item_name")
        assert item_in_cart.text == "Sauce Labs Backpack", "Товар не добавлен в корзину"
        
        # 6. Нажатие кнопки Checkout
        checkout_button = driver.find_element(By.ID, "checkout")
        checkout_button.click()
        time.sleep(2)  # Задержка после нажатия кнопки
        
        # Проверка перехода на страницу оформления
        WebDriverWait(driver, 10).until(
            EC.url_to_be("https://www.saucedemo.com/checkout-step-one.html")
        )
        
        # 7. Заполнение данных
        first_name = driver.find_element(By.ID, "first-name")
        last_name = driver.find_element(By.ID, "last-name")
        postal_code = driver.find_element(By.ID, "postal-code")
        continue_button = driver.find_element(By.ID, "continue")
        
        first_name.send_keys("John")
        last_name.send_keys("Doe")
        postal_code.send_keys("12345")
        continue_button.click()
        time.sleep(2)  # Задержка после нажатия кнопки
        
        # Проверка перехода на страницу подтверждения
        WebDriverWait(driver, 10).until(
            EC.url_to_be("https://www.saucedemo.com/checkout-step-two.html")
        )
        
        # 8. Нажатие кнопки Finish
        finish_button = driver.find_element(By.ID, "finish")
        finish_button.click()
        time.sleep(2)  # Задержка после нажатия кнопки
        
        # Проверка успешного оформления заказа
        WebDriverWait(driver, 10).until(
            EC.url_to_be("https://www.saucedemo.com/checkout-complete.html")
        )
        
        # Проверка сообщения об успешном оформлении
        complete_header = driver.find_element(By.CLASS_NAME, "complete-header")
        assert complete_header.text == "Thank you for your order!"
        
        print("Тест успешно пройден: покупка оформлена")
        
    except Exception as e:
        print(f"Тест не пройден: {str(e)}")
    finally:
        # Закрытие браузера
        driver.quit()

if __name__ == "__main__":
    test_successful_purchase()
```
